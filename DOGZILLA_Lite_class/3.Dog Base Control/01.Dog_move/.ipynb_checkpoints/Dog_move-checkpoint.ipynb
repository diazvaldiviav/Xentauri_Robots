```python
from xgolib import XGO
import time
import ipywidgets as widgets
from IPython.display import display
from ipywidgets import interact, widgets, Button, VBox, IntSlider

dog = XGO("xgolite")
```

```python
#设置步态 #Set gait
output00 = widgets.Output()

def on_button_clicked(b):
    with output00:
        print("Button clicked:", b.description)
    if b.description == 'trot':
        b.icon == 'check'
        dog.gait_type('trot')
    elif b.description == 'walk':
        b.icon == 'check'
        dog.gait_type('walk')
    elif b.description == 'high_walk':
        b.icon == 'check'
        dog.gait_type('high_walk')
    elif b.description == 'slow_trot':
        b.icon == 'check'
        dog.gait_type('slow_trot')


button_action_o1 = widgets.Button(        
    description='trot',     
    button_style='', # 'success', 'info', 'warning', 'danger' or ''     
    tooltip='Description',     
    icon='uncheck' )

button_action_o2 = widgets.Button(        
    description='walk',     
    button_style='', # 'success', 'info', 'warning', 'danger' or ''     
    tooltip='Description',     
    icon='uncheck' )

button_action_o3 = widgets.Button(        
    description='high_walk',     
    button_style='', # 'success', 'info', 'warning', 'danger' or ''     
    tooltip='Description',     
    icon='uncheck' )

button_action_o4 = widgets.Button(        
    description='slow_trot',     
    button_style='', # 'success', 'info', 'warning', 'danger' or ''     
    tooltip='Description',     
    icon='uncheck' )



button_action_o1.on_click(on_button_clicked)
button_action_o2.on_click(on_button_clicked)
button_action_o3.on_click(on_button_clicked)
button_action_o4.on_click(on_button_clicked)


box_btn_11 = widgets.HBox([button_action_o1, button_action_o2, button_action_o3, button_action_o4])
box_v = widgets.VBox([box_btn_11])
box_display = widgets.VBox([box_v, output00])
display(box_display)
```

```python
#设置频率 #Set frequency
output = widgets.Output()

def on_button_clicked(b):
    with output:
        print("Button clicked:", b.description)
    if b.description == 'slow':
        b.icon == 'check'
        dog.pace('slow')
    elif b.description == 'normal':
        dog.pace('normal')
    elif b.description == 'high':
        dog.pace('high')

button_action_1 = widgets.Button(        
    description='slow',     
    button_style='', # 'success', 'info', 'warning', 'danger' or ''     
    tooltip='Description',     
    icon='uncheck' )

button_action_2 = widgets.Button(        
    description='normal',     
    button_style='', # 'success', 'info', 'warning', 'danger' or ''     
    tooltip='Description',     
    icon='uncheck' )

button_action_3 = widgets.Button(        
    description='high',     
    button_style='', # 'success', 'info', 'warning', 'danger' or ''     
    tooltip='Description',     
    icon='uncheck' )

button_action_1.on_click(on_button_clicked)
button_action_2.on_click(on_button_clicked)
button_action_3.on_click(on_button_clicked)

box_btn_1 = widgets.HBox([button_action_1, button_action_2, button_action_3])
box_v = widgets.VBox([box_btn_1])
box_display = widgets.VBox([box_v, output])
display(box_display)
```

```python
#设置上下左右的步长 #Set the step size up, down, left, right
def dog_movex_sport(x): 
    dog.move('x',x)
    
def dog_movey_sport(y):    
    dog.move('y',y)



interact(dog_movex_sport,\
        x = widgets.IntSlider(min=-25,max=25,step=1,value=0))
interact(dog_movey_sport,\
        y = widgets.IntSlider(min=-18,max=18,step=1,value=0))
```
